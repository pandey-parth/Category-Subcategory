## Whatever columns you are using for searching, use fillna, otherwise can create problem

```python
df['descriptions']=df['descriptions'].fillna('')
df['skills']=df['skills'].fillna('')
df['course_title']=df['course_title'].fillna('')
```

```python
## If language column is present then only use below code and make new column 'text'
```

```python
df=df[df['language']=='English']
df['text']=df['course_title']+' '+df['descriptions']+' '+df['skills']
df
```

```python
df=df.reset_index()
df=df.drop(columns=['index'])
df['text']
```

```python
from whoosh.fields import Schema, TEXT, KEYWORD, ID, STORED
from whoosh.analysis import SimpleAnalyzer
from whoosh.analysis import RegexTokenizer
schema = Schema(path = ID(unique=True),
                text=TEXT(analyzer=SimpleAnalyzer(),stored=True))

import os, os.path
from whoosh import index

if not os.path.exists("indexdir20"):  # Can use any file name but use this only for simplicity
    os.mkdir("indexdir20")

idx = index.create_in("indexdir20", schema)
```

## Below code need not to be change

```python
subcateg=[]
category=[]
for i in range(len(df)):
    idx = index.open_dir("indexdir20")
    writer=idx.writer()

    writer.update_document(path=u"/a",text=df.loc[i,'text'])
    writer.commit()
    str=''
    str1=''
    str_arr=[]
    from whoosh.qparser import QueryParser
    with idx.searcher() as searcher:
        for j in subcategory:
            query = QueryParser("text", idx.schema).parse(j)
            results=searcher.search(query,limit=None)
            if(len(results)>0):
                if len(str)==0:
                    str+=j
                else:
                    str+=';'
                    str+=j
                if 'Business' not in str_arr:
                    for k in Business:
                        if k==j:
                            str_arr.append('Business')
                            if len(str1)==0:
                                str1+=j
                            else:
                                str1+=';'
                                str1+='Business'
                if 'Art and Design' not in str_arr:
                    for k in A_D:
                        if k==j:
                            str_arr.append('Art and Design')
                            if len(str1)==0:
                                str1+=j
                            else:
                                str1+=';'
                                str1+='Art and Design'
                if 'Computer Science' not in str_arr:
                    for k in cs:
                        if k==j:
                            str_arr.append('Computer Science')
                            if len(str1)==0:
                                str1+=j
                            else:
                                str1+=';'
                                str1+='Computer Science'
                if 'Data Science' not in str_arr:
                    for k in DS:
                        if k==j:
                            str_arr.append('Data Science')
                            if len(str1)==0:
                                str1+=j
                            else:
                                str1+=';'
                                str1+='Data Science'
                if 'Health and Fitness' not in str_arr:
                    for k in H_F:
                        if k==j:
                            str_arr.append('Health and Fitness')
                            if len(str1)==0:
                                str1+=j
                            else:
                                str1+=';'
                                str1+='Health and Fitness'
                if 'Language and Communication' not in str_arr:
                    for k in L_C:
                        if k==j:
                            str_arr.append('Language and Communication')
                            if len(str1)==0:
                                str1+=j
                            else:
                                str1+=';'
                                str1+='Language and Communication'
                if 'Life Science' not in str_arr:
                    for k in L_S:
                        if k==j:
                            str_arr.append('Life Science')
                            if len(str1)==0:
                                str1+=j
                            else:
                                str1+=';'
                                str1+='Life Science'
                if 'Marketing' not in str_arr:
                    for k in M:
                        if k==j:
                            str_arr.append('Marketing')
                            if len(str1)==0:
                                str1+=j
                            else:
                                str1+=';'
                                str1+='Marketing'
                if 'Math and Logic' not in str_arr:
                    for k in M_L:
                        if k==j:
                            str_arr.append('Math and Logic')
                            if len(str1)==0:
                                str1+=j
                            else:
                                str1+=';'
                                str1+='Math and Logic'
                if 'Physical Science and Engineering' not in str_arr:
                    for k in P_S:
                        if k==j:
                            str_arr.append('Physical Science and Engineering')
                            if len(str1)==0:
                                str1+=j
                            else:
                                str1+=';'
                                str1+='Physical Science and Engineering'
                if 'Personal Development' not in str_arr:
                    for k in P_L:
                        if k==j:
                            str_arr.append('Personal Development')
                            if len(str1)==0:
                                str1+=j
                            else:
                                str1+=';'
                                str1+='Personal Development'
                if 'Social Science' not in str_arr:
                    for k in S_S:
                        if k==j:
                            str_arr.append('Social Science')
                            if len(str1)==0:
                                str1+=j
                            else:
                                str1+=';'
                                str1+='Social Science'
                
                        
    subcateg.append(str)  
    category.append(str1)
```
